```python
# Funkcja do ekstrakcji tekstu z PDF
def extract_text_from_pdf(pdf_path, method='pdfplumber'):
    """
    WyciƒÖga tekst z pliku PDF.
    
    Args:
        pdf_path (str): ≈öcie≈ºka do pliku PDF
        method (str): 'pdfplumber' lub 'pypdf2'
    
    Returns:
        str: Wyekstraktowany tekst
    """
    
    if method == 'pdfplumber':
        with pdfplumber.open(pdf_path) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text()
            return text
    
    elif method == 'pypdf2':
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text()
            return text
    
    else:
        raise ValueError(f"Nieznana metoda: {method}. U≈ºyj 'pdfplumber' lub 'pypdf2'")


# Test funkcji
print("üß™ TEST FUNKCJI:")
print("=" * 50)

# Test z pdfplumber
result = extract_text_from_pdf(pdf_path, method='pdfplumber')
print(f"‚úÖ pdfplumber: {len(result)} znak√≥w")

# Test z PyPDF2
result = extract_text_from_pdf(pdf_path, method='pypdf2')
print(f"‚úÖ PyPDF2: {len(result)} znak√≥w")

print("\nüéâ Funkcja dzia≈Ça poprawnie!")
```

```python
# ULEPSZONA funkcja z obs≈ÇugƒÖ b≈Çƒôd√≥w
def extract_text_from_pdf(pdf_path, method='pdfplumber', clean=True):
    """
    WyciƒÖga tekst z pliku PDF z obs≈ÇugƒÖ b≈Çƒôd√≥w.
    
    Args:
        pdf_path (str): ≈öcie≈ºka do pliku PDF
        method (str): 'pdfplumber' lub 'pypdf2'
        clean (bool): Czyczy≈õciƒá tekst (usunƒÖƒá nadmiarowe spacje/newliny)
    
    Returns:
        str: Wyekstraktowany tekst lub None w przypadku b≈Çƒôdu
    """
    
    try:
        # Sprawd≈∫ czy plik istnieje
        if not os.path.exists(pdf_path):
            print(f"‚ùå B≈ÇƒÖd: Plik nie istnieje: {pdf_path}")
            return None
        
        # Ekstrakcja
        if method == 'pdfplumber':
            with pdfplumber.open(pdf_path) as pdf:
                text = ""
                for page in pdf.pages:
                    page_text = page.extract_text()
                    if page_text:  # Sprawd≈∫ czy strona ma tekst
                        text += page_text + "\n"
        
        elif method == 'pypdf2':
            with open(pdf_path, 'rb') as file:
                pdf_reader = PyPDF2.PdfReader(file)
                text = ""
                for page in pdf_reader.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + "\n"
        
        else:
            print(f"‚ùå B≈ÇƒÖd: Nieznana metoda '{method}'")
            return None
        
        # Czyszczenie tekstu (opcjonalne)
        if clean and text:
            # Usu≈Ñ nadmiarowe spacje
            text = ' '.join(text.split())
        
        return text
    
    except Exception as e:
        print(f"‚ùå B≈ÇƒÖd podczas ekstrakcji: {e}")
        return None


# Test ulepszonej funkcji
print("üß™ TEST ULEPSZONEJ FUNKCJI:")
print("=" * 50)

result = extract_text_from_pdf(pdf_path, method='pdfplumber', clean=True)
if result:
    print(f"‚úÖ Sukces! D≈Çugo≈õƒá: {len(result)} znak√≥w")
    print(f"üìÑ Fragment: {result[:200]}...")
else:
    print("‚ùå Ekstrakcja nie powiod≈Ça siƒô")
```

```python
# Funkcja por√≥wnujƒÖca obie metody
def compare_pdf_methods(pdf_path):
    """
    Por√≥wnuje PyPDF2 i pdfplumber dla danego PDF.
    Zwraca statystyki i rekomendacjƒô.
    
    Args:
        pdf_path (str): ≈öcie≈ºka do pliku PDF
    
    Returns:
        dict: Wyniki por√≥wnania
    """
    
    print("üîç POR√ìWNANIE METOD EKSTRAKCJI")
    print("=" * 60)
    
    # Test pdfplumber
    text_pdfplumber = extract_text_from_pdf(pdf_path, method='pdfplumber', clean=True)
    
    # Test PyPDF2
    text_pypdf2 = extract_text_from_pdf(pdf_path, method='pypdf2', clean=True)
    
    if not text_pdfplumber or not text_pypdf2:
        print("‚ùå Nie uda≈Ço siƒô wyekstraktowaƒá tekstu")
        return None
    
    # Statystyki
    results = {
        'pdfplumber': {
            'length': len(text_pdfplumber),
            'text': text_pdfplumber
        },
        'pypdf2': {
            'length': len(text_pypdf2),
            'text': text_pypdf2
        }
    }
    
    # Wy≈õwietl por√≥wnanie
    print(f"\nüìä STATYSTYKI:")
    print(f"{'Metoda':<15} {'D≈Çugo≈õƒá':<10} {'Ocena'}")
    print("-" * 60)
    print(f"{'pdfplumber':<15} {results['pdfplumber']['length']:<10} {'‚úÖ Zwykle lepsze formatowanie'}")
    print(f"{'PyPDF2':<15} {results['pypdf2']['length']:<10} {'‚ö†Ô∏è Prostsze, szybsze'}")
    
    # Rekomendacja
    print(f"\nüí° REKOMENDACJA:")
    if abs(results['pdfplumber']['length'] - results['pypdf2']['length']) < 100:
        print("   Obie metody dajƒÖ podobne wyniki - wybierz dowolnƒÖ")
        results['recommended'] = 'both'
    else:
        print("   pdfplumber - lepsze formatowanie dla z≈Ço≈ºonych PDF")
        results['recommended'] = 'pdfplumber'
    
    return results


# Test funkcji por√≥wnawczej
comparison = compare_pdf_methods(pdf_path)
```
