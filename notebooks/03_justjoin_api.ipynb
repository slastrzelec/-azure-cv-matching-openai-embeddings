```python
# ==============================
# FREE JOB SEARCH API - FULL CELL
# ==============================

# Je≈õli potrzeba odkomentuj:
# !pip install requests pandas

import requests
import pandas as pd
import os

# ==============================
# 1. KONFIGURACJA
# ==============================

API_KEY = "TU_WSTAW_SW√ìJ_API_KEY"

BASE_URL = "https://www.freejobapi.com/api/jobs"  # Sprawd≈∫ dok≈Çadny endpoint w dokumentacji API

HEADERS = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

QUERY = "data scientist"
LOCATION = "Poland"
PAGE = 1

OUTPUT_PATH = "jobs_raw.csv"


# ==============================
# 2. FUNKCJA POBIERANIA OFERT
# ==============================

def fetch_jobs(query: str, location: str, page: int = 1):
    
    params = {
        "query": query,
        "location": location,
        "page": page
    }
    
    try:
        response = requests.get(
            BASE_URL,
            headers=HEADERS,
            params=params,
            timeout=30
        )
        
        response.raise_for_status()
        data = response.json()
        
        jobs = data.get("jobs", [])
        print(f"Pobrano {len(jobs)} ofert.")
        
        return jobs
    
    except requests.exceptions.RequestException as e:
        print(f"B≈ÇƒÖd podczas pobierania danych: {e}")
        return []


# ==============================
# 3. WYWO≈ÅANIE API
# ==============================

jobs = fetch_jobs(QUERY, LOCATION, PAGE)

if not jobs:
    print("Brak danych do przetworzenia.")
else:
    
    # ==============================
    # 4. DATAFRAME
    # ==============================
    
    df = pd.DataFrame(jobs)
    
    # Wyb√≥r kluczowych kolumn (je≈õli istniejƒÖ)
    expected_columns = ["title", "company", "location", "description", "url"]
    available_columns = [col for col in expected_columns if col in df.columns]
    
    df_clean = df[available_columns]
    
    print("Dostƒôpne kolumny:")
    print(df_clean.columns.tolist())
    
    print("\nPodglƒÖd danych:")
    display(df_clean.head())
    
    # ==============================
    # 5. ZAPIS DO CSV
    # ==============================
    
    df_clean.to_csv(OUTPUT_PATH, index=False, encoding="utf-8")
    
    print(f"\nZapisano dane do pliku: {OUTPUT_PATH}")
```

```python
import requests
import pandas as pd

# ==============================
# CONFIG
# ==============================

BASE_URL = "https://www.themuse.com/api/public/jobs"

PARAMS = {
    "page": 1,
    "descending": True
}

OUTPUT_PATH = "jobs_global.csv"

# ==============================
# FETCH DATA
# ==============================

try:
    response = requests.get(BASE_URL, params=PARAMS, timeout=30)
    response.raise_for_status()
    
    data = response.json()
    jobs = data.get("results", [])
    
    print(f"Pobrano {len(jobs)} ofert (surowych).")
    
except requests.exceptions.RequestException as e:
    print(f"B≈ÇƒÖd API: {e}")
    jobs = []

# ==============================
# FILTER GLOBAL "DATA SCIENTIST"
# ==============================

filtered_jobs = []

for job in jobs:
    title = job.get("name", "").lower()
    description = job.get("contents", "").lower()
    
    if "data scientist" in title or "data scientist" in description:
        filtered_jobs.append({
            "title": job.get("name"),
            "company": job.get("company", {}).get("name"),
            "location": ", ".join([loc["name"] for loc in job.get("locations", [])]),
            "description": job.get("contents"),
            "url": job.get("refs", {}).get("landing_page")
        })

print(f"Po filtracji: {len(filtered_jobs)} ofert.")

# ==============================
# SAVE
# ==============================

if filtered_jobs:
    df = pd.DataFrame(filtered_jobs)
    display(df.head())
    
    df.to_csv(OUTPUT_PATH, index=False, encoding="utf-8")
    print(f"\nZapisano do {OUTPUT_PATH}")
else:
    print("Brak ofert po filtracji.")
```

```python
import requests
import json

print("üåç TEST The Muse API")
print("=" * 60)

# The Muse endpoint
BASE_URL = "https://www.themuse.com/api/public/jobs"

# Parametry wyszukiwania
params = {
    "category": "Data Science",  # Kategoria
    "page": 1,
    "descending": True,
    "page_size": 50  # Pobierz wiƒôcej (domy≈õlnie 20)
}

print(f"üì° Pobieranie ofert z The Muse...")

try:
    response = requests.get(BASE_URL, params=params, timeout=30)
    response.raise_for_status()
    
    data = response.json()
    all_jobs = data.get("results", [])
    
    print(f"‚úÖ Pobrano {len(all_jobs)} ofert")
    
    # FILTROWANIE - ML/Python keywords
    keywords = ['machine learning', 'ml engineer', 'python', 'pytorch', 
                'tensorflow', 'data scientist', 'ai', 'deep learning']
    
    filtered_jobs = []
    for job in all_jobs:
        title = job.get("name", "").lower()
        description = job.get("contents", "").lower()
        
        # Sprawd≈∫ czy zawiera keywords
        if any(kw in title or kw in description for kw in keywords):
            
            # Przygotuj strukturƒô zgodnƒÖ z naszym app.py
            filtered_jobs.append({
                "id": job.get("id"),
                "title": job.get("name"),
                "company": job.get("company", {}).get("name", "Unknown"),
                "description": job.get("contents", "")[:500],  # Ogranicz opis
                "requirements": ", ".join(job.get("tags", [])),  # Tagi jako requirements
                "location": ", ".join([loc["name"] for loc in job.get("locations", [])]),
                "salary": "N/A",  # The Muse nie zawsze ma salary
                "link": job.get("refs", {}).get("landing_page", "")
            })
    
    print(f"‚úÖ Po filtracji ML/Python: {len(filtered_jobs)} ofert")
    
    # Poka≈º przyk≈Çad
    if filtered_jobs:
        print(f"\nüìù Przyk≈Çadowe oferty:")
        for i, job in enumerate(filtered_jobs[:3], 1):
            print(f"\n{i}. {job['title']}")
            print(f"   üè¢ {job['company']}")
            print(f"   üìç {job['location']}")
            print(f"   üîó {job['link'][:50]}...")
        
        # Zapisz do JSON
        output_file = "data/muse_jobs.json"
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(filtered_jobs, f, indent=2, ensure_ascii=False)
        
        print(f"\n‚úÖ Zapisano {len(filtered_jobs)} ofert do {output_file}")
    else:
        print("‚ö†Ô∏è Brak ofert po filtracji")
        
except Exception as e:
    print(f"‚ùå B≈ÇƒÖd: {e}")
```

```python
import requests
import json

print("üåç The Muse API - WERSJA DZIA≈ÅAJƒÑCA")
print("=" * 60)

BASE_URL = "https://www.themuse.com/api/public/jobs"

# Kilka wyszukiwa≈Ñ dla r√≥≈ºnych keywords
search_terms = [
    "machine learning",
    "data scientist",
    "python developer",
    "AI engineer"
]

all_jobs = []

for term in search_terms:
    print(f"\nüîé Szukam: '{term}'...")
    
    try:
        params = {
            "search": term,
            "page": 0  # The Muse u≈ºywa 0-indexed
        }
        
        response = requests.get(BASE_URL, params=params, timeout=10)
        
        if response.status_code == 200:
            data = response.json()
            jobs = data.get("results", [])
            
            print(f"   ‚úÖ Znaleziono: {len(jobs)} ofert")
            
            # Dodaj do listy (unikaj duplikat√≥w po ID)
            existing_ids = {j.get("id") for j in all_jobs}
            for job in jobs:
                if job.get("id") not in existing_ids:
                    all_jobs.append(job)
        else:
            print(f"   ‚ùå B≈ÇƒÖd: {response.status_code}")
            
    except Exception as e:
        print(f"   ‚ùå B≈ÇƒÖd: {e}")

print(f"\n{'='*60}")
print(f"üìä PODSUMOWANIE: Pobrano {len(all_jobs)} unikalnych ofert")

# Przetw√≥rz do naszego formatu
formatted_jobs = []

for job in all_jobs:
    # WyciƒÖgnij lokalizacje
    locations = job.get("locations", [])
    location_str = ", ".join([loc.get("name", "") for loc in locations]) if locations else "Remote"
    
    # WyciƒÖgnij tagi jako requirements
    tags = job.get("tags", [])
    requirements = ", ".join([tag.get("name", "") for tag in tags]) if tags else "Not specified"
    
    formatted_jobs.append({
        "id": job.get("id"),
        "title": job.get("name", "Unknown"),
        "company": job.get("company", {}).get("name", "Unknown"),
        "description": job.get("contents", "")[:1000],  # Ogranicz do 1000 znak√≥w
        "requirements": requirements,
        "location": location_str,
        "salary": "Not specified",  # The Muse rzadko ma salary
        "link": job.get("refs", {}).get("landing_page", "")
    })

print(f"‚úÖ Przetworzono {len(formatted_jobs)} ofert")

# Poka≈º przyk≈Çady
if formatted_jobs:
    print(f"\nüìù Przyk≈Çadowe oferty (TOP 5):")
    for i, job in enumerate(formatted_jobs[:5], 1):
        print(f"\n{i}. {job['title']}")
        print(f"   üè¢ {job['company']}")
        print(f"   üìç {job['location']}")
        print(f"   üîß {job['requirements'][:60]}...")
    
    # Zapisz do JSON
    output_file = "data/muse_jobs.json"
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(formatted_jobs, f, indent=2, ensure_ascii=False)
    
    print(f"\n‚úÖ Zapisano {len(formatted_jobs)} ofert do {output_file}")
    print(f"üí° Mo≈ºesz teraz u≈ºyƒá tego pliku zamiast sample_jobs.json w app.py")
else:
    print("‚ö†Ô∏è Brak ofert")
```
